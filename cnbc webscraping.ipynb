```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get ("https://www.cnbc.com/world/?region=world")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
header = []
for i in soup.find_all('span',class_="nav-menu-buttonText"):
    header.append(i.text)

    
header  
```
