```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver = webdriver.Chrome(r'C:\Users\prasad kalekar\Selenium\chromedriver.exe')
driver.maximize_window()
driver = webdriver.Chrome('chromedriver.exe')
url = ' https://www.forbes.com/?sh=605023e82254'

driver.get(url)
```

```python
name = []
b = driver.find_elements(By.XPATH,'//div[@class="personName"]')
for i in b:
    name.append(i.text)
name 
```

```python
net = []
w = driver.find_elements(By.XPATH,'//div[@class="netWorth"]')
for i in w:
    net.append(i.text)
net 
```

```python
country = []
A = driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]')
for i in A:
    country.append(i.text)
country 
```

```python
com = []
A = driver.find_elements(By.XPATH,'//span[@class="source-text"]')
for i in A:
    com.append(i.text)
com 
```

```python
age = []
A = driver.find_elements(By.XPATH,'//div[@class="age"]')
for i in A:
    age.append(i.text)
age 
```

```python
B = pd.DataFrame()
B['NAME'] = name
B['AGE'] = age
B['COUNTRY']=country
B['INDUSTRY'] = com
B['NET WORTH']=net
B
```
